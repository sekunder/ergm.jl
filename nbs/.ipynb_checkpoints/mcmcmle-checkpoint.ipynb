# Fit Edge-Density ERGM to Ground-Truth

```julia
using ergm.sampler
using ergm.spaces
using ergm.stats
using ergm.models
using ergm.inference
using ergm.optim
using Random

n = 10
stats = DeltaStats(
    function(G)
        sum(G.adjacency) / (n * (n - 1))
    end,
    function(G, s, u)
        i, x = u
        [s[1] + (x - G[i]) / (n * (n - 1))]
    end
)
θ0 = (i -> sign(i) * 10 ^ abs(i))(rand(-2:0.1:2))
m = ExponentialFamily(stats, [θ0])
G0 = DiGraph(rand(Bool, (n, n)))
s = GibbsSampler(
    G0,
    m,
    10,
    10
)
Gs, ss = sample(s, 1000)
#o = SGD(10.0, [0.0], 1000, 0.0, "../logs/a.txt")
o = ADAM(1.0, 0.9, 0.999, 1e-8, [0.0], 100, 0.0, "../logs/a.txt")
θs, Ls = mcmc_mle(Gs, m, o, 100, 10, 10)
```

```julia
using Gadfly
using DataFrames

df = DataFrame(s = Int[], θ = Float64[], r = String[])

a(θ) = exp(θ / (n * (n - 1))) / (1 + exp(θ / (n * (n - 1))))

for (s, θ) ∈ enumerate(θs)
    push!(df, (s, a(θ), "fit"))
    push!(df, (s, a(θ0), "target"))
end

Gadfly.with_theme(:dark) do
    plot(df, x=:s, y=:θ, color=:r, Geom.line)
end
```

```julia
using Gadfly
using DataFrames

df2 = DataFrame(s = Int[], L = Float64[])

for (s, L) ∈ enumerate(Ls)
    push!(df2, (s, L))
end

Gadfly.with_theme(:dark) do
    plot(df2, x=:s, y=:L, Geom.line)
end
```
